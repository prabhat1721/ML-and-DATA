### Gaussian Transformation
Some machine learning algos like LinearRegression and LogisticRegression assumes that features are normally distributed
##### logarithmic, reciprocal, squareroot, exponential, boxcox transformation


```python
import scipy.stats as stat
import pylab
```

```python
def plot_data(df, feature):
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    df[feature].hist()
    plt.subplot(1, 2, 2)
    stat.probplot(df[feature], dist='norm', plot=pylab)
    plt.show()
```

```python
plot_data(df, 'Age')
```

```python
import numpy as np
df['Age_log'] = np.log(df['Age'])
plot_data(df, 'Age_log')
```

```python
df['Age_reciprocal'] = 1/df.Age
plot_data(df, 'Age_reciprocal')
```

```python
df['Age_sqrt'] = df.Age ** (1/2)
plot_data(df, 'Age_sqrt')
```

```python
df['Age_exp'] = df.Age ** (1/1.2)
plot_data(df, 'Age_exp')
```

### BoxCot 
t(y) = (y exp(lambda)-1)/lambda
y:response variable
lambda:transfromation parameter(-5 to 5)



```python
df['Age_Boxcox'], parameters = stat.boxcox(df['Age'])
```

```python
print(parameters)
```

```python
plot_data(df, 'Age_Boxcox')
```
