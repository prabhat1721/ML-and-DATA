```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. Simple plot with 4 numbers

```python
plt.plot([1, 3, 2,  4])
plt.show()
```

### 2. Points have x and y values; add title and axis labels

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Test Plot', fontsize=12, color='g')
plt.xlabel('number n')
plt.ylabel('n^2')
plt.show()
```

### 3. Change figsize, plot red dots; set axis scales x:0-6 and y:0-20

```python
plt.figure(figsize=(2, 5)) # 2 inches wide x 5 inches tall
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'rx') # red-o
plt.axis([0, 6, 0, 20]) # [xmin, xmax, ymin, ymax]
plt.annotate('square it', (3, 6))
plt.show()
```

### Bar chart with four bars

```python
plt.clf()  # clear figure
x = np.arange(4)
y = [8.8, 5.2, 3.6, 5.9]
plt.xticks(x, ('Ankit', 'Hans', 'Joe', 'Flaco'))
#plt.bar(x, y)
#plt.bar(x, y, color='lime') 
plt.bar(x, y, color=['lime', 'r', 'tan', 'yellow'])
plt.show()
```

### Two sets of 10 random dots plotted

```python
d = {'Red o': np.random.rand(10),
      'Grn x': np.random.rand(10)}
df = pd.DataFrame(d)
df.plot(style=['ro', 'gx'])
plt.show()
```

### Time Series - six months of random floats

```python
ts = pd.Series(np.random.randn(180), index=pd.date_range('1/1/2018', periods=180))
df = pd.DataFrame(np.random.randn(180, 3), index=ts.index, columns=list('ABC'))
df.cumsum().plot()
plt.show()
```

### Random dots in a scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = (30*np.random.rand(N))**2   # 0 to 15 point radii
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7)
plt.show()
```

### Styling, Jitter

```python
plt.figure(figsize=(8,6))
sns.stripplot(x='weight', y='origin', data=cars,
             hue='cylinders',
             palette='bright',
             dodge=True, size=10,
             jitter=0.5,
             alpha=0.5);
```

```python
from ipywidgets import interactive

def jitter_widget(jitter=0.1):
    sns.stripplot(x='weight', y='origin', data=cars,
             hue='cylinders',
             palette='bright',
             dodge=True, size=6,
             jitter=0.5);
```

```python
interactive(jitter_widget, jitter=(0, 0.65, 0.05))
```

```python
sns.stripplot(x='weight', y='origin', data=cars,
             hue='cylinders',
             palette='bright',
             dodge=True, size=7,
             jitter=0.5,
             alpha=0.5,
             linewidth=1,
             marker='*');
```
