```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
cancer_df = sklearn.datasets.load_breast_cancer()
cancer_df
```

```python
df = pd.DataFrame(cancer_df.data, columns = cancer_df.feature_names)
df.head()
```

```python
df['label'] = cancer_df.target
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['label'].value_counts()
```

```python
X = df.drop(columns='label', axis=1)
y = df['label']
```

```python
X
```

```python
X = np.asarray(X)
y = np.asarray(y)
```

### GridSearchCV

```python
model = SVC()
```

```python
parameters = {
              'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
               'C':[1, 5, 10, 20]
}
```

```python
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X, y)
```

```python
classifier.cv_results_
```

```python
best_params = classifier.best_params_
best_params
```

```python
highest_accuracy = classifier.best_score_
highest_accuracy
```

```python
result = pd.DataFrame(classifier.cv_results_)
result
```

```python
grid_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
grid_search_result
```

### RandomizedSearchCV


```python
model = SVC()
```

```python
parameters = {
              'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
               'C':[1, 5, 10, 20]
}
```

```python
classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X, y)
```

```python
classifier.cv_results_
```

```python
best_params = classifier.best_params_
best_params
```

```python
highest_accuracy = classifier.best_score_
highest_accuracy
```

```python
result = pd.DataFrame(classifier.cv_results_)
result
```

```python
randomized_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
randomized_search_result
```
